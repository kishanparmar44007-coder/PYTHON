pivoting

```python
data={
    'date':['20244-01-01','2024-01-01','2024-01-02','2024-01-02'],
    'product':['laptop','mouse','laptop','mouse'],
    'sales':[1200,100,1300,120]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.pivot(index='date',columns='product',values='sales')
```

```python
df.pivot(index="product",columns="date",values="sales")
```

```python
df.pivot_table(index="product",columns="date",values="sales",aggfunc="sum")
```
