```python
df1=pd.DataFrame({
    "ID":[1,2,3,4],
    "Name":["Ram","Sita","Jagu","Dholu"],
    "Category":["A","B","A","A"]
})
```

```python
df2=pd.DataFrame({
    "ID":[3,4,5,6],
    "Salary":[25000,12000,21000,11000]
})
```

# Merge

- inner  = only matching
- outer  = all merge
- left   = left all   & right only matching
- right  = right all  & left only matching

```python
df1
```

```python
df2
```

```python
inner=pd.merge(df1,df2,how="inner",on="ID")
inner
```

```python
left=pd.merge(df1,df2,how="left",on="ID")
left
```

```python
right=pd.merge(df1,df2,how="right",on="ID")
right
```

```python
outer=pd.merge(df1,df2,how="outer",on="ID")
outer
```

# Join

- join using index no column names

```python
d1=df1.set_index("ID")
d1
```

```python
d2=df2.set_index("ID")
d2
```

```python
d1.join(d2,how="inner")
```

```python
d1.join(d2,how="left")
```

```python
d1.join(d2,how="right")
```

```python
d1.join(d2,how="outer")
```

- You use merge when you want full control.
- You use join when both DataFrames already have matching index.
