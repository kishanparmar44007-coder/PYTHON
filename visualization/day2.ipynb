```python
import matplotlib.pyplot as plt
```

stack chart

```python
days=['mon', 'tue', 'wen', 'thu', 'fri']

product_a=[5, 7, 6, 4, 5]
product_b=[3, 5, 2, 6, 4]
product_c=[2, 3, 4, 3, 2]

plt.bar(days, product_a, label='product_A')

plt.bar(days, product_b, label='product_B')

plt.bar(days, product_c, label='product_C')

plt.legend()

plt.show()
```

```python
plt.bar(days, product_a, label='product_A', width=0.6)

plt.bar(days, product_b, label='product_B', width=0.3)

plt.bar(days, product_c, label='product_C', width=0.4)

plt.legend()

plt.show()
```

```python
days=['mon', 'tue', 'wen', 'thu', 'fri']

product_a=[5, 7, 6, 4, 5]
product_b=[3, 5, 2, 6, 4]
product_c=[2, 3, 4, 3, 2]

plt.barh(days, product_a, label='product_A')
plt.barh(days, product_b, label='product_B')
plt.barh(days, product_c, label='product_C')

plt.legend()

plt.show()
```

```python
# STACK AREA CAHRT (MULTIPALE LINE PLOTS WITH FILL =>)

plt.stackplot(
    days,
    product_a,
    product_b,
    product_c,
    labels=['product_A','product_B', 'product_C']
    )

plt.legend()

plt.grid(axis='y', alpha=0.6)

plt.show()
```

SUBPLOT

```python
X=[1, 2, 3, 4, 5]

y1=[10, 20, 30, 40, 50]

x1=[5, 15, 25, 35, 45]

fig,axs=plt.subplots(1, 2, figsize=(10, 4))

axs[0].plot(x1, y1, color='red')
axs[0].set_title('line plot')

axs[1].bar(x1, y1, color='blue')
axs[1].set_title('bar chart')

plt.tight_layout()

plt.show()
```

```python
x=[1, 2, 3, 4, 5]

y=[2, 4, 6, 8, 10]

fig,axs=plt.subplots(2, 2, figsize=(10, 6))

axs[0,0].plot(x,y)
axs[0,0].set_title('plot-1')

axs[0,1].bar(x,y)
axs[0,1].set_title('bar-1')

axs[1,0].barh(x,y)
axs[1,0].set_title('scatter-1')

axs[1,1].barh(x,y)
axs[1,1].set_title('hrozontal bar')

plt.tight_layout()

plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
```

```python

x=[1, 2, 3, 4, 5]

y1=[10, 20, 30, 40, 50]

x1=[5, 15, 25, 35, 45]

plt.plot(x, y1, label='line 1')
plt.plot(x, x1, label='line 2')

plt.title('lengend example')

plt.xlabel('x-axis')

plt.ylabel('y-axis')

plt.legend(loc='upper right')

plt.legend(fontsize=15)

plt.legend(title='lines')

plt.legend(frameon=False)
```
