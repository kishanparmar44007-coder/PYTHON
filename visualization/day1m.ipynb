LINE CHART

marker value : 0 , * , +

```python
month=['jan','feb','mar','jun','apr','dec','nov','jan','fab','mar','sep','aug']

revenue=[200,500,300,150,250,560,190,360,580,891,198,200]

plt.plot(month, revenue, markersize=8, color='black', markerfacecolor='red')

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.plot(month, revenue, marker=0, markersize=5, color='black', markerfacecolor='blue',  linewidth=3)

for i, val in enumerate(revenue):

    plt.text(i, val + 120, str(val), ha='center', fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.9, axis='both')

plt.savefig("monthly_revenue.png", dpi=300)

plt.show()
```

```python
math=[78,85,90,88,76,95,89,84,91,87]

scinec=[82,79,88,90,85,92,80,86,89,91]

plt.hist(math, bins=5, alpha=0.5, label='math', color='red')

plt.hist(scinec, bins=5, label='scinec', color='tomato')

plt.legend()

plt.title('comparison of math and scince')

plt.xlabel('score range')

plt.ylabel('frequency')

plt.show()
```

```python
sales=[200,250,300,280,350]

month=['jan', 'fab', 'mar', 'apr', 'may']

plt.plot(month, sales, marker=0)

plt.title('monthly sales')

plt.xlabel('month')

plt.ylabel('sales')

plt.grid(True)

plt.show()
```

pie chart (donut chart)

```python
sales=[50,30,20]

labels=['laptop', 'mobiles', 'tablets']

colors=['red', 'orange', 'pink']

plt.pie(sales, labels=labels, autopct='%1.1f%%', colors=colors)

center_color=plt.Circle((0,0), 0.6, color='white')

fig=plt.gcf()

fig.gca().add_artist(center_color)

plt.title('sales distrbution')

plt.show()
```

SCATTER PLOT

```python
size=[20,40,60,80,100]

weigth=[30,60,90,120,150]

plt.scatter(size, weigth, s=100, c='blue')

plt.xlabel('size')

plt.ylabel('weigth')

plt.title('size vs weigth')

plt.show()
```
